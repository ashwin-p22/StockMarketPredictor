```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
```

```python
sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y="Close", use_index=True)
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc_model = RandomForestClassifier(n_estimators = 100, min_samples_split = 100, random_state = 1)

train_set = sp500.iloc[:-100]
test_set = sp500.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
rfc_model.fit(train_set[predictors], train_set["Target"])
```

```python
from sklearn.metrics import precision_score

preds = rfc_model.predict(test_set[predictors])
```

```python
import pandas as pds
preds = pds.Series(preds, index = test_set.index)
```

```python
precision_score(test_set["Target"], preds)
```

```python
combined = pds.concat([test_set["Target"], preds], axis = 1)
```

```python
combined.plot()
```

```python
def predict(train_set, test_set, predictors, rfc_model):
    rfc_model.fit(train_set[predictors], train_set["Target"])
    preds = rfc_model.predict(test_set[predictors])
    preds = pds.Series(preds, index = test_set.index, name = "Predictions")
    combined = pds.concat([test_set["Target"], preds], axis = 1)
    return combined
```

```python
def backtest(data, rfc_model, predictors, start = 2500, step = 250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train_set = data.iloc[0:i].copy()
        test_set = data.iloc[i:(i+step)].copy()
        predictions = predict(train_set, test_set, predictors, rfc_model)
        all_predictions.append(predictions)
    return pds.concat(all_predictions)
```

```python
predictions = backtest(sp500, rfc_model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    moving_averages = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / moving_averages["Close"]

    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna()
```

```python
sp500
```

```python
rfc_model = RandomForestClassifier(n_estimators = 200, min_samples_split = 50, random_state = 1)
```

```python
def predict(train_set, test_set, predictors, rfc_model):
    rfc_model.fit(train_set[predictors], train_set["Target"])
    preds = rfc_model.predict_proba(test_set[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pds.Series(preds, index = test_set.index, name = "Predictions")
    combined = pds.concat([test_set["Target"], preds], axis = 1)
    return combined
```

```python
predictions = backtest(sp500, rfc_model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```
